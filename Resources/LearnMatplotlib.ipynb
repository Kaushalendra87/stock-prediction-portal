```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Types of plot

## Line plots

```python
x = [1,2,3,4,5]
y=[10,20,25,30,40]
```

```python
plt.plot(x,y);
plt.title('Simple Line Plot');
plt.xlabel("X-axis label");
plt.ylabel("Y-axis label");     
plt.grid(True)
```

## Bar Chart

```python
months = ['Jan','Feb' , 'March', 'April']
revenue = [100,200,300,400]
```

```python
plt.figure(figsize=(10,6))
plt.bar(months,revenue);
plt.title('Simple Line Plot');
plt.xlabel("Months");
plt.ylabel("Revenue");  
```

## Scatter Plot

## Subplots and Legends

```python
days = [ 1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [ 8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();
```

## Subplots

```python
## First subplot - Original Price
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.title('Original prices over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

## Seocond subplot - Predicted Price
plt.subplot(3,1,2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Predicted prices over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

## Third subplot - Differences between Original & Predicted Prices
plt.subplot(3,1,3)
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.bar(days, price_difference, color='purple', label='Price Difference')
plt.title('Differences between Original & Predicted Prices');
plt.xlabel('Days');
plt.ylabel('Price Difference');
plt.legend();
plt.tight_layout()
```

## Saving the figure

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();
plt.savefig('example_prediction.png', dpi=300) #Save the image
```
